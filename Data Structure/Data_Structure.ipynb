## ðŸ‡ºðŸ‡¸ **Linked List**

### **1. Definition**

A linked list is a linear data structure, in which the elements are not stored at contiguous memory locations. The elements in a linked list are linked using pointers as shown in the below image:
Singlelinkedlist.png

**Node Structure**: A node in a linked list typically consists of two components:

**Data**: It holds the actual value or data associated with the node.

**Next Pointer**: It stores the memory address (reference) of the next node in the sequence.

**Head and Tail**: The linked list is accessed through the head node, which points to the first node in the list. The last node in the list points to NULL or nullptr, indicating the end of the list. This node is known as the tail node.



### **2. Why LinkedList?**

Here are a few advantages of a linked list that is listed below, it will help you understand why it is necessary to know.

*  **Dynamic Data structure**: The size of memory can be allocated or de-allocated at run time based on the operation insertion or deletion.
*  **Ease of Insertion/Deletion**: The insertion and deletion of elements are simpler than arrays since no elements need to be shifted after insertion and deletion, Just the address needed to be updated.
*  **Efficient Memory Utilization**: As we know Linked List is a dynamic data structure the size increases or decreases as per the requirement so this avoids the wastage of memory.
*  **Implementation**: Various advanced data structures can be implemented using a linked list like a stack, queue, graph, hash maps, etc.

**Example**:

In a system, if we maintain a sorted list of IDs in an array id[] = [1000, 1010, 1050, 2000, 2040].

If we want to insert a new ID 1005, then to maintain the sorted order, we have to move all the elements after 1000 (excluding 1000).

Deletion is also expensive with arrays until unless some special techniques are used. For example, to delete 1010 in id[], everything after 1010 has to be moved due to this so much work is being done which affects the efficiency of the code.

## ðŸ‡ºðŸ‡¸ **Stack**

### **1. Definition**

Stack is a linear data structure that follows a particular order in which the operations are performed. The order may be LIFO(Last In First Out) or FILO(First In Last Out). LIFO implies that the element that is inserted last, comes out first and FILO implies that the element that is inserted first, comes out last.

stack.drawio2.png

There are many real-life examples of a stack. Consider an example of plates stacked over one another in the canteen. The plate which is at the top is the first one to be removed, i.e. the plate which has been placed at the bottommost position remains in the stack for the longest period of time. So, it can be simply seen to follow LIFO(Last In First Out)/FILO(First In Last Out) order.


### **2. Why Stack?**

**Advantages of Stack**:

* Easy implementation: Stack data structure is easy to implement using arrays or linked lists, and its operations are simple to understand and implement.

* Efficient memory utilization: Stack uses a contiguous block of memory, making it more efficient in memory utilization as compared to other data structures.

* Fast access time: Stack data structure provides fast access time for adding and removing elements as the elements are added and removed from the top of the stack.

* Helps in function calls: Stack data structure is used to store function calls and their states, which helps in the efficient implementation of recursive function calls.

* Supports backtracking: Stack data structure supports backtracking algorithms, which are used in problem-solving to explore all possible solutions by storing the previous states.

* Used in Compiler Design: Stack data structure is used in compiler design for parsing and syntax analysis of programming languages.

* Enables undo/redo operations: Stack data structure is used to enable undo and redo operations in various applications like text editors, graphic design tools, and software development environments.

**Applications of Stack**:
* CD/DVD stand.

* Stack of books in a book shop.

* Call center systems.

* Undo and Redo mechanism in text editors.

* The history of a web browser is stored in the form of a stack.

* Call logs, E-mails, and Google photos in any gallery are also stored in form of a stack.

* YouTube downloads and Notifications are also shown in LIFO format(the latest appears first ).

* Allocation of memory by an operating system while executing a process.


### **3. Implementation & Operations**

##### **Structure**

* class **Stack**

##### **Basic Operations**

the **Stack** class provides the following methods:

* __init__: Initializes an empty stack using a list.

* **is_empty**: Checks if the stack is empty. Returns True if the stack is empty, otherwise False.

* **push**: Adds an item to the top of the stack.

* **pop**: Removes and returns the item at the top of the stack. If the stack is empty, it returns an appropriate error message.

* **search**: Search an element and return its position

* **peek**: Returns the item at the top of the stack without removing it. If the stack is empty, it returns an appropriate error message.

* **size**: Returns the number of elements in the stack.

##### **Original Implementation**

````python
class Stack:
    def __init__(self):
        self.stack = []

    def is_empty(self):
        return len(self.stack) == 0

    def push(self, item):
        self.stack.append(item)

    def pop(self):
        if self.is_empty():
            return "Stack is empty. Cannot pop an item."
        return self.stack.pop()

    def search(self, item):
        position = len(self.stack) - 1

        while position >= 0:
            if self.stack[position] == item:
                return position
            position -= 1

    def peek(self):
        if self.is_empty():
            return "Stack is empty. No item to peek."
        return self.stack[-1]

    def size(self):
        return len(self.stack)

stack = Stack()

print(stack.is_empty())  # Output: True

stack.push(10)
stack.push(20)
stack.push(30)

print(stack.size())  # Output: 3
print(stack.peek())  # Output: 30

item = stack.pop()
print(item)  # Output: 30

print(stack.size())  # Output: 2
print(stack.search(10))  # Output: 2 (item found)
print(stack.search(50))  # Output: -1 (item not found)
````

##### **Implementation using Linked List**

````python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class Stack:
    def __init__(self):
        self.head = None
        self.size = 0

    def is_empty(self):
        return self.head is None

    def push(self, item):
        new_node = Node(item)
        new_node.next = self.head
        self.head = new_node
        self.size += 1

    def pop(self):
        if self.is_empty():
            return "Stack is empty. Cannot pop an item."
        popped_item = self.head.data
        self.head = self.head.next
        self.size -= 1
        return popped_item

    def search(self, item):
        position = 0
        current = self.head

        while current:
            if current.data == item:
                return position
            position += 1
            current = current.next

        return -1

    def peek(self):
        if self.is_empty():
            return "Stack is empty. No item to peek."
        return self.head.data

    def get_size(self):
        return self.size

stack = Stack()

print(stack.is_empty())  # Output: True

stack.push(10)
stack.push(20)
stack.push(30)

print(stack.search(10))  # Output: 2 (item found at the index 2 of the linkedlist 30 -> 20 -> 10)


print(stack.get_size())  # Output: 3
print(stack.peek())  # Output: 30

item = stack.pop()
print(item)  # Output: 30

print(stack.get_size())  # Output: 2

print(stack.search(50))  # Output: -1 (item not found)
````

## ðŸ‡ºðŸ‡¸ **Queue**

### **1. Definition**

A Queue is defined as a linear data structure that is open at both ends and the operations are performed in First In First Out (FIFO) order.


Queue-Data-Structures.png

### **2. Why Queue?**

* Queue can handle multiple data.

* We can access both ends.

* They are fast and flexible.

### **3. Implementation & Operations**

##### **Structure**
* class **Queue**

##### **Basic Operations**

The **Queue** class provides the following methods:

* __init__: Initializes an empty queue using a list.

* **is_empty**: Checks if the queue is empty. Returns True if the queue is empty, otherwise False.

* **enqueue**: Adds an item to the rear of the queue using the append method of the list.

* **dequeue**: Removes and returns the item at the front of the queue using the pop method with an index of 0 to remove the first item.

* **search**: Search the item and return its position

* **peek**: Returns the item at the front of the queue without removing it. If the queue is empty, it returns an appropriate error message.

* **size**: Returns the number of elements in the queue.

##### **Original Implementation**


````python
class Queue:
    def __init__(self):
        self.queue = []

    def is_empty(self):
        return len(self.queue) == 0

    def enqueue(self, item):
        self.queue.append(item)

    def dequeue(self):
        if self.is_empty():
            return "Queue is empty. Cannot dequeue an item."
        return self.queue.pop(0)

    def peek(self):
        if self.is_empty():
            return "Queue is empty. No item to peek."
        return self.queue[0]

    def size(self):
        return len(self.queue)

    def search(self, item):
        for index, value in enumerate(self.queue):
            if value == item:
                return index
        return -1

queue = Queue()

print(queue.is_empty())  # Output: True

queue.enqueue(10)
queue.enqueue(20)
queue.enqueue(30)

print(queue.size())  # Output: 3
print(queue.peek())  # Output: 10

item = queue.dequeue()
print(item)  # Output: 10

print(queue.size())  # Output: 2

````

##### **Implementation using Linked List**


````python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


class Queue:
    def __init__(self):
        self.front = None
        self.rear = None

    def is_empty(self):
        return self.front is None

    def enqueue(self, item):
        new_node = Node(item)
        if self.is_empty():
            self.front = new_node
            self.rear = new_node
        else:
            self.rear.next = new_node
            self.rear = new_node

    def dequeue(self):
        if self.is_empty():
            return "Queue is empty. Cannot dequeue an item."
        removed_item = self.front.data
        if self.front == self.rear:
            self.front = None
            self.rear = None
        else:
            self.front = self.front.next
        return removed_item

    def peek(self):
        if self.is_empty():
            return "Queue is empty. No item to peek."
        return self.front.data

    def size(self):
        count = 0
        current = self.front
        while current:
            count += 1
            current = current.next
        return count

    def search(self, item):
        position = 0
        current = self.front

        while current:
            if current.data == item:
                return position
            position += 1
            current = current.next

        return -1

queue = Queue()

print(queue.is_empty())  # Output: True

queue.enqueue(10)
queue.enqueue(20)
queue.enqueue(30)

print(queue.size())  # Output: 3
print(queue.peek())  # Output: 10

item = queue.dequeue()
print(item)  # Output: 10

print(queue.size())  # Output: 2

print(queue.search(30))  # Output: 1 (index of 30 relative to the front)
print(queue.search(50))  # Output: -1 (item not found)
````

## ðŸ‡ºðŸ‡¸ **Binary Tree**

### **1. Definition**

Binary Tree is defined as a tree data structure where each node has at most 2 children. Since each element in a binary tree can have only 2 children, we typically name them the left and right child.

### **2. Why Binary Tree?**

Binary trees are an important data structure in computer science and have various applications. Here are some reasons why binary trees are needed:

Decision trees find applications in machine learning, data mining, and pattern recognition. 5. **Huffman Coding**: Binary trees are fundamental in Huffman coding, a popular compression algorithm. Huffman trees are used to assign variable-length codes to characters based on their frequency of occurrence, resulting in efficient data compression.

6. **Binary Search and Range Queries**: Binary trees support efficient binary search operations, which can locate elements in a sorted list or array in logarithmic time complexity. They are also used for range queries, where elements within a specific range need to be retrieved or processed.

Binary trees provide a flexible and efficient way to organize and process data, making them a fundamental data structure in computer science. They enable efficient searching, sorting, expression evaluation, decision making, compression, and other operations in various domains. ### **3. Implementation and Operations**

##### **Structure**


A **Binary Tree** is represented by **a pointer to the topmost node** (commonly known as the **â€œrootâ€**) of the tree. If the tree is empty, then the value of the root is NULL. Each node of a Binary Tree contains the following parts:

* **Data**
* **Pointer to left child**
* **Pointer to right child**

##### **Basic Operations**


* **Inserting** an element.
* **Removing** an element.
* **Searching** for an element.
* **Traversing** the tree (pre, in, post - order).

##### **Auxiliary Operations**


* Finding the **height** of the tree
* Find the **level** of a node of the tree
* Finding the **size** of the entire tree.

##### **Implementation**

````python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

class BinaryTree:
    def __init__(self):
        self.root = None

    def insert(self, data):
        if self.root is None:
            self.root = Node(data)
        else:
            self._insert_recursive(self.root, data)

    def _insert_recursive(self, current, data):
        if data < current.data:
            if current.left is None:
                current.left = Node(data)
            else:
                self._insert_recursive(current.left, data)
        elif data > current.data:
            if current.right is None:
                current.right = Node(data)
            else:
                self._insert_recursive(current.right, data)

    def remove(self, data):
        self.root = self._remove_recursive(self.root, data)

    def _remove_recursive(self, current, data):
        if current is None:
            return current

        if data < current.data:
            current.left = self._remove_recursive(current.left, data)
        elif data > current.data:
            current.right = self._remove_recursive(current.right, data)
        else:
            if current.left is None:
                return current.right
            elif current.right is None:
                return current.left
            else:
                min_node = self._find_min_node(current.right)
                current.data = min_node.data
                current.right = self._remove_recursive(current.right, min_node.data)

        return current

    def _find_min_node(self, current):
        while current.left is not None:
            current = current.left
        return current

    def search(self, data):
        return self._search_recursive(self.root, data)

    def _search_recursive(self, current, data):
        if current is None or current.data == data:
            return current

        if data < current.data:
            return self._search_recursive(current.left, data)
        else:
            return self._search_recursive(current.right, data)

    def traverse_inorder(self):
        self._traverse_inorder_recursive(self.root)
        print()

    def _traverse_inorder_recursive(self, current):
        if current is not None:
            self._traverse_inorder_recursive(current.left)
            print(current.data, end=" ")
            self._traverse_inorder_recursive(current.right)

    def get_height(self):
        return self._get_height_recursive(self.root)

    def _get_height_recursive(self, current):
        if current is None:
            return 0

        left_height = self._get_height_recursive(current.left)
        right_height = self._get_height_recursive(current.right)

        return max(left_height, right_height) + 1

    def get_level(self, data):
        return self._get_level_recursive(self.root, data, 1)

    def _get_level_recursive(self, current, data, level):
        if current is None:
            return 0

        if current.data == data:
            return level

        left_level = self._get_level_recursive(current.left, data, level + 1)
        if left_level != 0:
            return left_level

        right_level = self._get_level_recursive(current.right, data, level + 1)
        return right_level

    def get_size(self):
        return self._get_size_recursive(self.root)

    def _get_size_recursive(self, current):
        if current is None:
            return 0

        left_size = self._get_size_recursive(current.left)
        right_size = self._get_size_recursive(current.right)

        return left_size + right_size + 1
````

### **2. Why Min Heap?**

A min-heap is a useful data structure that is often used in various algorithms and applications. Here are a few reasons why a min-heap is needed:

1. **Efficient Retrieval of Minimum Element**: The primary purpose of a min-heap is to efficiently retrieve the minimum (smallest) element from a collection of elements. The minimum element is always at the root of the min-heap, making it easy to access in constant time. This property is valuable in scenarios where you frequently need to retrieve the minimum element, such as priority queue implementations.

2. **Efficient Sorting**: Min-heaps are commonly used in sorting algorithms like heapsort. The heap data structure allows for an efficient sorting process by repeatedly extracting the minimum element from the heap and building a sorted sequence. Heapsort has a time complexity of O(n log n), making it an efficient sorting algorithm.

3. **Priority Queue**: Min-heaps are widely used to implement priority queues, where elements have different levels of priority. The priority queue ensures that the element with the highest priority (smallest value in the case of a min-heap) is always at the front and can be efficiently retrieved. Priority queues find applications in various domains like task scheduling, graph algorithms (e.g., Dijkstra's algorithm), and more.

4. **Efficient Update Operations**: Min-heaps provide efficient support for operations like insertions and updates. When inserting an element into a min-heap, it can be placed at the appropriate position in the heap and then heapified to maintain the heap property. Similarly, when updating an element, it can be modified and then heapified to restore the heap property. These operations have a time complexity of O(log n), making them efficient even for large heaps.

5. **Efficient Selection of Top-k Elements**: Min-heaps can be used to efficiently select the top-k smallest elements from a large collection. By maintaining a min-heap of size k, you can iterate over the elements and replace the largest element in the heap if a smaller element is encountered. At the end of the process, the min-heap will contain the top-k smallest elements.

Overall, a min-heap is a valuable data structure that offers efficient access to the minimum element, allows for efficient sorting and selection operations, and finds application in priority queues and various algorithms.

### **3. Implementation & Operations**

##### **Structure**
Class **MinHeap** with attribute **heap**: A list that stores the elements of the heap.

##### **Basic Operations**
* **parent**(index): Returns the index of the parent node for the given index.
* **left_child**(index): Returns the index of the left child node for the given index.
* **right_child**(index): Returns the index of the right child node for the given index.
* **swap**(i, j): Swaps the elements at indices i and j in the heap.
* **insert**(item): Inserts an item into the heap and maintains the heap property by performing the heapify_up operation.
* **heapify_up**(index): Restores the heap property by moving the element at the given index up the heap if necessary.
* **extract_min**(): Removes and returns the minimum element (root) from the heap and maintains the heap property by performing the heapify_down operation.
* **heapify_down**(index): Restores the heap property by moving the element at the given index down the heap if necessary.

##### **Implementation**


````python
class MinHeap:
    def __init__(self):
        self.heap = []

    def parent(self, index):
        return (index - 1) // 2

    def left_child(self, index):
        return 2 * index + 1

    def right_child(self, index):
        return 2 * index + 2

    def swap(self, i, j):
        self.heap[i], self.heap[j] = self.heap[j], self.heap[i]

    def insert(self, item):
        self.heap.append(item)
        self.heapify_up(len(self.heap) - 1)

    def heapify_up(self, index):
        while index > 0 and self.heap[index] < self.heap[self.parent(index)]:
            parent_index = self.parent(index)
            self.swap(index, parent_index)
            index = parent_index

    def extract_min(self):
        if len(self.heap) == 0:
            return "Heap is empty. Cannot extract minimum element."
        min_item = self.heap[0]
        last_item = self.heap.pop()
        if len(self.heap) > 0:
            self.heap[0] = last_item
            self.heapify_down(0)
        return min_item

    def heapify_down(self, index):
        while True:
            left_child_index = self.left_child(index)
            right_child_index = self.right_child(index)
            smallest_index = index

            if left_child_index < len(self.heap) and self.heap[left_child_index] < self.heap[smallest_index]:
                smallest_index = left_child_index

            if right_child_index < len(self.heap) and self.heap[right_child_index] < self.heap[smallest_index]:
                smallest_index = right_child_index

            if smallest_index == index:
                break

            self.swap(index, smallest_index)
            index = smallest_index

heap = MinHeap()
heap.insert(10)
heap.insert(5)
heap.insert(15)
heap.insert(2)

print(heap.extract_min())  # Output: 2
print(heap.extract_min())  # Output: 5
````
